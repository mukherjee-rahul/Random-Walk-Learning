```python
#Defining the constants

N = 100 # total no. of steps
# Do not go beyond 10000000

P = 0.5 # probality of taking right
Q = 1-P #probability of taking left
```

```python
#importing libraries for matrix and table operations and plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating a table of 1 * N+1
Table = np.zeros(N+1)
print(Table)
print('Shape of the TAble is: ')
print(Table.shape)
```

```python
#This commands generates a random float between 0 and 1
np.random.rand()
```

```python

fl = 0
fr = 0
for i in range(0,N):
    #Generating a random number
    random_number = np.random.rand()
    #checking if the random_number is less than or equal to P
    if random_number<P:
        fr+=1
        Table[i+1] = Table[i]+1 # adding 1 to the i th element to get (i+1)th element of the table
    else:
        fl+=1
        Table[i+1] = Table[i]-1 # else subtracting 1
        
#Print final position i.e. the last element of the Table
print(Table[N])
print('No. of left steps:',fl,'No. of right steps:',fr)

plt.plot(Table)
```

Mean position for M ensembles of random walks of 100 steps

The below code just encapsulates the above code into a function which returns the last step only

```python
def random_walk():
    fl = 0
    fr = 0
    for i in range(0,N):
        #Generating a random number
        random_number = np.random.rand()

        #checking if the random_number is less than or equal to P
        if random_number<P:
            fr+=1
            
        else:
            fl+=1
            
    # outputs the final position = (number of righ steps - number of left steps)
    return fr-fl

#Call the function
random_walk()
```

```python
#
M=1000
sum_final = 0
for i in range(M):
    sum_final = sum_final + random_walk()
#Taking average
sum_final = sum_final/M
sum_final
```
